```python
# prompt: ETL

# ETL process: Join relevant dataframes for a more complete view of orders
df_orders_customers = df_orders.join(df_customers, on='customer_id', how='inner')
df_orders_customers_items = df_orders_customers.join(df_items, on='order_id', how='inner')
df_orders_customers_items_products = df_orders_customers_items.join(df_products, on='product_id', how='inner')
df_full = df_orders_customers_items_products.join(df_payments, on='order_id', how='inner')

# Select relevant columns for analysis and drop potential duplicates or less useful ones
columns_to_keep = [
    'order_id', 'customer_id', 'order_status', 'order_purchase_timestamp',
    'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date',
    'order_estimated_delivery_date', 'customer_unique_id', 'customer_zip_code_prefix',
    'customer_city', 'customer_state', 'order_item_id', 'product_id', 'seller_id',
    'shipping_limit_date', 'price', 'freight_value', 'product_category_name',
    'product_name_lenght', 'product_description_lenght', 'product_photos_qty',
    'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm',
    'payment_type', 'payment_installments', 'payment_value'
]
df_cleaned = df_full.select(columns_to_keep)

# Handle missing values (example: dropping rows with nulls in critical columns)
# You can choose other strategies like imputation based on the data and task
df_cleaned = df_cleaned.dropna(subset=['order_id', 'customer_id', 'product_id', 'seller_id', 'payment_value'])

# Data type conversion (example: converting date strings to timestamp)
from pyspark.sql.functions import to_timestamp

date_columns = [
    'order_purchase_timestamp', 'order_approved_at', 'order_delivered_carrier_date',
    'order_delivered_customer_date', 'order_estimated_delivery_date', 'shipping_limit_date'
]
for col_name in date_columns:
    # Using 'yyyy-MM-dd HH:mm:ss' as a common format, adjust if necessary
    df_cleaned = df_cleaned.withColumn(col_name, to_timestamp(col_name))

# Feature Engineering (example: calculating order duration)
from pyspark.sql.functions import datediff, col

df_cleaned = df_cleaned.withColumn(
    'order_duration_days',
    datediff(col('order_delivered_customer_date'), col('order_purchase_timestamp'))
)

# Show a sample of the cleaned and transformed data
df_cleaned.show(5)

# Print schema to check data types
df_cleaned.printSchema()
```

```python
# prompt: df cleaned k means

from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.evaluation import ClusteringEvaluator

# Select numerical features for K-Means
numerical_features = [
    'price', 'freight_value', 'product_name_lenght',
    'product_description_lenght', 'product_photos_qty', 'product_weight_g',
    'product_length_cm', 'product_height_cm', 'product_width_cm',
    'payment_installments', 'payment_value', 'order_duration_days'
]

# Handle potential nulls in selected numerical columns by filling with a reasonable value (e.g., 0 or mean)
# For simplicity, let's fill with 0. A more robust approach might involve mean or median imputation.
for col_name in numerical_features:
    df_cleaned = df_cleaned.fillna(0, subset=[col_name])


# Assemble features into a single vector column
assembler = VectorAssembler(inputCols=numerical_features, outputCol="features")
df_kmeans = assembler.transform(df_cleaned)

# Select relevant features for clustering and drop rows with nulls if any remain
# (VectorAssembler handles nulls by default by setting the corresponding value to 0 in the vector,
# but it's good practice to ensure the input columns are handled beforehand)
df_kmeans = df_kmeans.select("features").dropna()

# Define the number of clusters (k)
k = 3  # You can adjust this based on domain knowledge or by using methods like the Elbow method or Silhouette score

# Create a KMeans model
kmeans = KMeans().setK(k).setSeed(1)

# Train the model
model = kmeans.fit(df_kmeans)

# Make predictions
predictions = model.transform(df_kmeans)

# Evaluate clustering by computing Silhouette score
evaluator = ClusteringEvaluator()

silhouette = evaluator.evaluate(predictions)
print(f"Silhouette with squared euclidean distance = {silhouette}")

# Shows the result.
centers = model.clusterCenters()



# You can join the cluster predictions back to the original cleaned dataframe if needed
# Note that joining back might require careful handling of potential duplicate rows
# introduced during previous joins, or operating on a distinct set of customer/order IDs for clustering.
# For this example, we've clustered the feature vectors derived from the cleaned dataframe.
```

```python
print("Cluster Centers: ",centers)
```
